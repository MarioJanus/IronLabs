```python
import numpy as np
import pandas as pd
employees=pd.read_csv(r"C:\Users\Mario\Documents\Labs\Module2\Lab_5_ Pivot-Tables-and-Crosstab\Employee.csv")
```

```python
employees.pivot_table
```

```python
employees.groupby(['Department']).agg({'Salary' :'mean'})
```

```python
employees.pivot_table(index=["Department"], values=["Salary"])
```

```python
employees.pivot_table(index=['Title'], columns=['Years'], values=['Salary'])
```

```python
employees.pivot_table(index=['Title'], columns=['Years'], values=['Salary'], aggfunc=np.mean)
```

```python
employees.pivot_table(index=['Department',  'Title'], values=['Salary'])
```

```python
employees.groupby(['Department']).agg({'Salary' :'mean', 'Department':'count'})
```

```python
aggfunc=('median', 'min', 'max', 'std')
employees.groupby(['Department','Title']).agg({'Salary' :aggfunc})
```

```python
employees.groupby(['Department','Title']).agg({'Salary' :aggfunc}).fillna(0)
```

```python
employees.groupby(['Department','Title']).agg({'Salary' :aggfunc}).fillna(0)
```

```python
employees.groupby(['Department']).agg(NumbersEmployee=('Department', 'min'))
```
