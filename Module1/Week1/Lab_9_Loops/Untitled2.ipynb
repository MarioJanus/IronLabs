```python
import random
sample_list_1 = random.sample(range(1, 100), 80)
#print(sample_list_1)
set1 = set(sample_list_1)
print("Set1",set1)

sample_list_2=[]
for i in range(1,80):
    n=random.randint(1,100)
    sample_list_2.append(n)
#print(sample_list_2)

set2 = set(sample_list_2)
print(len(set2))
print("Set2",set2)


sample_list_3=[]
for i in set1:
    if i not in set2:
        sample_list_3.append(i)
set3=set(sample_list_3)
print("Set3",set3)

sample_list_4=[]
for i in set2:
    if i not in set1:
        sample_list_4.append(i)
set4=set(sample_list_4)
print("Set4",set4)

sample_list_5=[]
for i in set2:
    if i in set2:
        sample_list_5.append(i)
set5=set(sample_list_5)
print("Set5",set5)

set6=set()
set6.update(set3)
set6.update(set5)
print("Set6",set6)

if set1==set6:
    print("Set1 and Set6 are equel")
else:
    print("Set1 and Set6 are not equal")
    
print(set2.issubset(set1))
print(set3.issubset(set1))
set_union345=set3.union(set4,set5)
print(set_union345)
set_union12=set1.union(set2)
print(set_union345)
if set_union345==set_union12:
    print("Merged set3,4,5 and merged set1,2 are equal")
else:
    print("Merged set3,4,5 and merged set1,2 are not equal")

set1.pop()
print(set1)

list_to_remove = [1, 9, 11, 19, 21, 29, 31, 39, 41, 49, 51, 59, 61, 69, 71, 79, 81, 89, 91, 99]
for i in list_to_remove:
    if i in set1:
        set1.discard(i)
print(set1)
```
