```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_excel(r"C:\Users\Mario\Documents\IronMario\Module1\Week3\Project_Week3\2 -Entrepreneurial competency in university students.xlsx")
```

```python
data.describe()
```

```python
data.rename(columns={'Target IndividualProject ':'TargetIndividualProject'}, inplace=True)
```

```python
data.rename(columns={'Target-ent_competency':'TargetEntCompetency'}, inplace=True)
```

```python
to_lower_letter=['TargetIndividualProject','Gender','City','Influenced','MentalDisorder','ReasonsForLack','ReasonsForLack']
for i in to_lower_letter:
       data[i]=data[i].str.lower()
```

```python
data.loc[data['KeyTraits']== "Rrresilience", 'KeyTraits']= "resilience"
data
```

```python
col_fill = ['Perseverance', 'DesireToTakeInitiative',
        'Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']
for i in col_fill:
    data.loc[data[i].isna()==True,i]=3
data
```

```python
data.loc[data['MentalDisorder'].isna()==True,'MentalDisorder']='undisclosed'
```

```python
data.loc[data['Age'].isna()==True,'Age']=round(data['Age'].mean())
data['Age'].mean()
```

```python
cols_to_int = ['Age', 'Perseverance', 'DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']
data[cols_to_int] = data[cols_to_int].astype(int)
data
```

```python
data.drop(columns="ReasonsForLack", inplace=True)
data
```

```python
data.to_csv(r'C:\Users\Mario\Documents\IronMario\Module1\Week3\Project_Week3\data_cleaned.csv') 
```
