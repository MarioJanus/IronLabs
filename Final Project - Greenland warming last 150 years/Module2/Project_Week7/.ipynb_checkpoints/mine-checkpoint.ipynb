```python
import numpy as np
import pandas as pd
df=pd.read_excel(r"C:\Users\Mario\Documents\Labs\Module2\Project_Week7\Sample data set employability__Tableau_temp.xlsx")
df.head(3)
```

```python
df['Days since graduation'] = df['Days since graduation'].astype(int)
df['Portfolio'] = df['Portfolio'].astype(int)
df.head(3)
```

```python
df['Soft skills pct']=df['Soft skills pct'].map(lambda n: '{:,.2%}'.format(n))
df['Job ready pct']=df['Job ready pct'].map(lambda n: '{:,.2%}'.format(n))
df['Final project pct']=df['Final project pct'].map(lambda n: '{:,.2%}'.format(n))

df['Soft skills score']=df['Soft skills score'].map(lambda n: '{:,.3}'.format(n))
df.head(15)
```

```python
df.head(15)
```

```python
df.isnull().sum()
```

```python
df['Track']=df['Track'].fillna("WD") 
df['at risk']=df['at risk'].fillna(0)
df.loc[(df['at risk']!=0),"at risk"]=1
```
